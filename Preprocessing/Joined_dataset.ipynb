```python
import pandas as pd

# Define correct UNSW column names manually
col_names= [
    'Source IP', 'Source Port', 'Destination IP', 'Destination Port',
    'Protocol', 'Connection State', 'Duration',
    'Source Bytes', 'Destination Bytes',
    'Source TTL', 'Destination TTL',
    'Source Packet Loss', 'Destination Packet Loss',
    'Service', 'Source Load', 'Destination Load',
    'Source Packets', 'Destination Packets',
    'Source Window Size', 'Destination Window Size',
    'Source TCP Base Seq', 'Destination TCP Base Seq',
    'Source Mean Packet Size', 'Destination Mean Packet Size',
    'Transaction Depth', 'Response Body Length',
    'Source Jitter', 'Destination Jitter',
    'Start Time', 'End Time',
    'Source Inter-Packet Arrival Time', 'Destination Inter-Packet Arrival Time',
    'TCP Round Trip Time', 'SYN-ACK Time', 'ACK Data Time',
    'Same IPs/Ports Flag', 'Count State TTL',
    'HTTP Method Count', 'FTP Login Flag', 'FTP Command Count',
    'Connections from Same Source to Same Service',
    'Connections to Same Destination Service',
    'Connections to Same Destination in Last Minute',
    'Connections from Same Source in Last Minute',
    'Connections from Same Source to Same Destination Port in Last Minute',
    'Connections from Same Destination Source Port in Last Minute',
    'Connections between Same Destination and Source in Last Minute',
    'Attack Category', 'Label'
]

# Load with correct column names
df1 = pd.read_csv("UNSW-NB15_1.csv", header=None, names=col_names, encoding="latin1", low_memory=False)
df2 = pd.read_csv("UNSW-NB15_2.csv", header=None, names=col_names, encoding="latin1", low_memory=False)
df3 = pd.read_csv("UNSW-NB15_3.csv", header=None, names=col_names, encoding="latin1", low_memory=False)
df4 = pd.read_csv("UNSW-NB15_4.csv", header=None, names=col_names, encoding="latin1", low_memory=False)

# Merge all
df = pd.concat([df1, df2, df3, df4], ignore_index=True)

print(df.shape)
print(df.columns)
print(df['Label'].value_counts())
print(df['Attack Category'].value_counts())
```

```python
from google.colab import files

# Save dataframe to CSV
df.to_csv("joined_dataset.csv", index=False)

# Download to your local system
files.download("joined_dataset.csv")
```
